```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_Car.csv')
```

```python
df.info()
```

```python
#graficas del modelo
'''
odometer
size
year
title_status
condition_encoding
cylinders_int
'''
```

```python
#Grafica Price vs Odometer
x = df['odometer'].sort_values()
y = df['price'].sort_values()
plt.title('Price vs Odometer')
plt.plot(x,y)
plt.xlabel('Odometer')
plt.ylabel('Price')
plt.show()
```

```python
#Grafica Price vs Size
x = df['size'].sort_values()
y = df['price'].sort_values()
plt.plot(x,y)
plt.xlabel('size')
plt.ylabel('price')
plt.title('Price vs Size')
plt.show()
```

```python
#Grafica Price vs year
x = df['year'].sort_values()
y = df['price'].sort_values()
plt.plot(x,y)
plt.xlabel('year')
plt.ylabel('price')
plt.title('Price vs Year')
plt.show()
```

```python
#Grafica Price vs title status
x = df['title_status'].sort_values()
y = df['price'].sort_values()
plt.plot(x,y)
plt.xlabel('title_status')
plt.ylabel('price')
plt.title('Price vs Title status')
plt.show()
```

```python
#Grafica Price vs condition_encoding
x = df['condition_encoding'].sort_values()
y = df['price'].sort_values()
plt.plot(x,y)
plt.xlabel('condition_encoding')
plt.ylabel('price')
plt.title('Price vs condition_encoding')
plt.show()
```

```python
#Grafica Price vs cylinders_int
x = df['cylinders_int'].sort_values()
y = df['price'].sort_values()
plt.plot(x,y)
plt.xlabel('cylinders_int ')
plt.ylabel('price')
plt.title('Price vs cylinders_int ')
plt.show()
```
